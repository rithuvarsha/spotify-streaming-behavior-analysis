```python
import pandas as pd
import json
import glob
import os
```

```python
file_pattern = os.path.join("D:\Rithu\Data Analytics\Portfolio Project\Spotify Extended Streaming History\data","Streaming_History_Audio_*.json")
files = glob.glob(file_pattern)
```

```python
dfs = []
for file in files:
    with open(file, 'r', encoding= 'utf-8') as f:
        data = pd.read_json(f)
        dfs.append(data)
```

```python
df = pd.concat(dfs, ignore_index = True)
```

```python
cols_to_drop=['episode_name','episode_show_name','spotify_episode_uri','audiobook_title','audiobook_uri','audiobook_chapter_uri','audiobook_chapter_title','conn_country','ip_addr']
df = df.drop(columns=cols_to_drop)
```

```python
df['ts']=pd.to_datetime(df['ts'])
```

```python
df['year']=df['ts'].dt.year
df['month']=df['ts'].dt.month_name()
df['day_of_week']=df['ts'].dt.day_name()
df['hour']=df['ts'].dt.hour
```

```python
df = df.rename(columns={
    'master_metadata_track_name':'track',
    'master_metadata_album_artist_name':'artist',
    'master_metadata_album_album_name':'album',
    'reason_start':'start_reason',
    'reason_end':'end_reason'
})
```

```python
df.to_csv("spotify_master_data.csv", index=False)
print(f"Success! Processed {len(df)} rows. Saved to 'spotify_master_data.csv'.")
```

```python
df['platform'] = df['platform'].str.lower()

def categorize_device(platform):
    if pd.isna(platform):
        return "Unknown"
    elif 'windows' in platform or 'osx' in platform or 'mac' in platform or 'desktop' in platform:
        return 'PC/Desktop'
    elif 'android' in platform or 'ios' in platform or 'iphone' in platform:
        return 'Mobile'
    elif 'amazon' in platform or 'bose' in platform or 'alexa' in platform or 'speaker' in platform:
        return 'Smart Speaker'
    else:
        return 'Other'

df['device_type'] = df['platform'].apply(categorize_device)
```

```python
print(df['device_type'].value_counts())
```

```python
df = df[df['ms_played'] >= 30000]
```

```python
df = df.dropna(subset=['track', 'album','artist'])
```

```python
df
```

```python
df['listening_minutes'] = df['ms_played'] / 60000
```

```python
def get_day_part(hour):
    if 5 <= hour < 12: return 'Morning'
    elif 12 <= hour < 17: return 'Afternoon'
    elif 17 <= hour < 21: return 'Evening'
    else: return 'Night'

df['day_part'] = df['hour'].apply(get_day_part)
```

```python
def get_season(month):
    if month in ['December', 'January', 'February']: return 'Winter'
    elif month in ['March', 'April', 'May']: return 'Spring'
    elif month in ['June', 'July', 'August']: return 'Summer'
    else: return 'Autumn'

df['season'] = df['month'].apply(get_season)
```

```python
df['is_weekend'] = df['day_of_week'].isin(['Saturday', 'Sunday'])
```

```python
df.to_csv('spotify_cleaned.csv', index=False)
```
